## Objectifs

- Comprendre les spécificités de Python ;
- Introduire les expressions, les opérateurs basiques et leurs priorités ;
- Comprendre ce qu'est une variable, son affectation et les types de variables que l'on retrouve ;
- Appréhender les fonctions de base qui permette une communication avec l'utilisateur d'un programme ;
- Comprendre les opérateurs de comparaison ;
- Introduire les opérateurs booléens ;
- Apprendre à implémenter une instruction conditionnelle ;
- Maîtrîser l'implémentation des boucles `for` ;
- Maîtriser l'implémentation des boucles `while` ;
- Créer une structure de données de type `list` ;
- Utiliser une structure de répétition dans une liste ;
- Manipuler les éléménts d'une liste ;
- Etre capable d'ouvrir et de fermer un fichier texte ;
- Savoir écrire dans un fichier texte ;
- Utiliser les différentes méthodes de lecture d'un fichier texte.
