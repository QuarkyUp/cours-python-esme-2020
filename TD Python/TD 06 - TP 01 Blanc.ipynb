L'indice de masse corporelle (IMC) est une mesure qui a été définie en 1997 par l'OMS (Organisation Mondiale de la Santé) et qui permet d'estimer la corpulence d'une personne dans le but d'évaluer les risques liés au surpoids chez l'adulte. Des intervalles standards (Dénutrition, Maigreur, Corpulence normale, Surpoids, Obésité) ont été définis en se basant sur la relation constatée statistiquement entre l'IMC et le taux de mortalité. Ces intervalles sont donnés dans le tableau suivant :

**Valeur de l'IMC**       | **Interprétation (selon l'OMS)**
---------  | ---------
Inférieur à 16| Anorexie ou dénutrition
Entre 16,5 et 18,5     | Insuffisance pondérale ou maigreur 
Entre 18,5 et 25| Corpulence normale
Entre 25 et 30    | Surpoids
Entre 30 et 35     | Obésité modérée
Entre 35 et 40     | Obésité élevée
Supérieur à 40    | Obésité morbide

Selon la classification de l'OMS, l'IMC est calculé comme suit : 
<center>IMC = masse / taille²</center>

La masse est donnée en kg et la taille en m.
